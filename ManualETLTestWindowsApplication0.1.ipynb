```python
from PyQt6.QtWidgets import *
from PyQt6.QtCore import *
from PyQt6.QtGui import * 
import os, sys
import subprocess as sp

app = QCoreApplication.instance()
if app is None:
    app = QApplication(sys.argv)

class Window(QWidget):
    def __init__(self):
        
        with open(r'config\CVsStage.csv', 'r') as cv_config:
            optionsstage = cv_config.read().split(',')
        with open(r'config\CVsProd.csv', 'r') as cv_config:
            optionsprod = cv_config.read().split(',')

        ListExtS = ['All', 'Exit', 'Prod']
        ListExtP = ['All', 'Exit', 'Stage']
        
        buttondesc = "Import Selected CV's"

        optionsstage.extend(ListExtS)
        optionsprod.extend(ListExtP)
        
        super().__init__()
        self.setWindowTitle("Testing Functionality")
        self.setWindowIcon(QIcon("PwC.jpg"))
        self.setGeometry(500, 200, 500, 400)
        
        layout = QVBoxLayout()
        self.setLayout(layout)
        
        self.label1 = QLabel("Select Prod CV's")
        self.label1.setAlignment(Qt.AlignmentFlag.AlignCenter)
        self.label1.adjustSize()
        layout.addWidget(self.label1)
        
        self.ProdSelect1 = QListWidget()
        self.ProdSelect1.setSelectionMode(QAbstractItemView.SelectionMode.ExtendedSelection)
        self.ProdSelect1.addItems(optionsstage)
        layout.addWidget(self.ProdSelect1)
        
        button1 = QPushButton(buttondesc)
        button1.clicked.connect(self.btn1clicked)
        layout.addWidget(button1)

        self.label2 = QLabel("Select Stage CV's")
        self.label2.setAlignment(Qt.AlignmentFlag.AlignCenter)
        self.label2.adjustSize()
        layout.addWidget(self.label2)
        
        self.StageSelect2 = QListWidget()
        self.StageSelect2.setSelectionMode(QAbstractItemView.SelectionMode.ExtendedSelection)
        self.StageSelect2.addItems(optionsprod)
        layout.addWidget(self.StageSelect2)
        
        button2 = QPushButton(buttondesc)
        button2.clicked.connect(self.btn2clicked)
        layout.addWidget(button2)

        #self.combo.setAlignment(Qt.AlignmentFlag.AlignLeft)

    def btn1clicked(self):
        self.label1.setText("New and Updated Text")
        
    def btn2clicked(self):
        self.label2.setText("New and Updated Text")
        
    def get(self):
        print(self.label1.text())

window = Window()
window.show()
app.exec()
```

```python
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import smtplib, ssl

SMTPHOST = "smtp.gmail.com"
SMTPPORT = 587
FROMEMAIL = "grason.moye@pwc.com"
FROMEMAILPASSWORD = "Qpwoeiruty10!!"

def CreateEmailMessage(toAddress, message):
    messageDetails = MIMEMultipart()
    
    messageDetails['From'] = FROMEMAIL
    messageDetails['To'] = toAddress
    messageDetails['Subject'] = "Test Email!"
    
    messageDetails.attach(MIMEText(message, 'plain'))
    
    return messageDetails

def SendEmailMessage(toAddress, message):
    messageDetails = CreateEmailMessage(toAddress, message)
    context = ssl.create_default_context()
    
    try:
        server = smtplib.SMTP(SMTPHOST, SMTPPORT)
        server.starttls(context=context)
        server.login(FROMEMAIL, FROMEMAILPASSWORD)
        server.sendmail(FROMEMAIL, toAddress, messageDetails.as_string())
    except Exception as ex:
        print(ex)
    finally:
        server.quit()
            
SendEmailMessage("grason.moye@pwc.com", "Hello, this is an official PwC test email! :)")
```
